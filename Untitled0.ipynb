```python

Answer 1:
      Effect of the Number of Neighbors in the Isomap Algorithm

In Isomap, the number of neighbors (denoted as k) affects how the local structure of the data is captured when constructing the neighborhood graph. The number of neighbors has a significant impact on the quality of the low-dimensional embedding:

When the number of neighbors is very large:

Global structure dominates: The algorithm becomes more similar to classical Multidimensional Scaling (MDS), as every point is connected to a large portion of the data. This can lead to losing fine-grained local structure, especially for data that lies on a highly non-linear manifold.

Risk of over-smoothing: Isomap might no longer preserve local geometries accurately, which could result in a poor low-dimensional representation of non-linear relationships.


When the number of neighbors is very low:

Disconnected graph: If too few neighbors are used, the graph might become disconnected, where some points have no neighbors at all. This can lead to poor estimation of the geodesic distances, which makes the low-dimensional embedding incorrect.

Overfitting local structure: The algorithm might over-emphasize local variations and fail to capture the global structure of the data.



Thus, choosing an appropriate number of neighbors is critical. A balance is required where both local and global structures are adequately represented.

Answer 2:
       2. When Isomap Is Superior to PCA

Isomap is often superior to PCA when the data lies on a non-linear manifold. PCA assumes that the principal components lie in a linear subspace, which can lead to poor dimensionality reduction if the data has non-linear structures.

Non-linear structures: If the data has curved manifolds or non-linear relationships, PCA may fail to capture the true underlying structure. Isomap, on the other hand, preserves geodesic distances and can uncover these non-linear relationships.

Geodesic distance: Isomap computes geodesic distances (i.e., shortest paths along the manifold), which helps it retain the true structure of data when projected into lower dimensions. This makes it preferable to PCA in scenarios where non-linear dimensionality reduction is needed.
Answer 3:
        3. Other Manifold Learning Methods and Key Differences

Some other popular manifold learning methods besides Isomap include:

Locally Linear Embedding (LLE):

Key difference: LLE constructs local linear approximations of the manifold and then finds a global embedding that preserves these local structures. Unlike Isomap, LLE doesnâ€™t rely on geodesic distances but rather minimizes reconstruction errors from linear approximations.

Advantages: Works well with manifolds with complicated global structure, as long as local linearity holds.


t-Distributed Stochastic Neighbor Embedding (t-SNE):

Key difference: t-SNE focuses on preserving local similarities by using a probabilistic approach to match points in high-dimensional space with points in the low-dimensional embedding. It uses a cost function based on Kullback-Leibler divergence.

Advantages: Very effective at visualizing clusters and patterns in high-dimensional data, especially for datasets with complex, non-linear manifolds.


Uniform Manifold Approximation and Projection (UMAP):

Key difference: UMAP is similar to t-SNE but more efficient and scalable. It uses both local and global structures to create a faithful lower-dimensional representation.

Advantages: It has better computational efficiency and maintains both local and some global structure, making it suitable for large datasets
Answer 4:
     4. Dealing with Missing Data in Manifold Learning

Missing data can pose a challenge to manifold learning, as many algorithms rely on distance metrics, and missing values can interfere with distance computations. Some strategies to handle missing data include:

Imputation:

Use techniques like mean imputation, k-nearest neighbors imputation, or more sophisticated methods like matrix factorization to fill in missing values before applying manifold learning.


Pairwise distance calculation:

For algorithms that depend on distance metrics (like Isomap), calculate pairwise distances only between observed (non-missing) values, though this may introduce some inaccuracies.


Probabilistic modeling:

Use probabilistic models, such as Gaussian processes or mixture models, to estimate missing values and propagate uncertainties into the manifold learning process.


Manifold learning with missing data:

Some methods, such as GP-LVM (Gaussian Process Latent Variable Models), inherently handle missing data by placing a probabilistic distribution over the latent variables. These methods can estimate a lower-dimensional representation even with incomplete observations.



These methods help mitigate the impact of missing data while still preserving the meaningful structure of the data for dimensionality reduction tasks.
```

### **`BAR-GRAPHS:-`**

Customization that is available for the Bar Chart â€“

**color:** For the bar faces

**edgecolor:** Color of edges of the bar

**linewidth:** Width of the bar edges

**width:** Width of the bar


```python
import matplotlib.pyplot as plt
import pandas as pd
names=['KIET','KIEK','KIEW']
scores=[92.87,72.9,70]
plt.bar(names,scores,color='green', linewidth=5, edgecolor='red')
plt.xlabel('\n names', size=20)
plt.ylabel('scores',size=20)
plt.show()
```

**`Visualize the data in Piechart`**:-

To visualize the data in pie chart we use pie()


```python
names=['pannu','prend','frienduuu']
scores=[92.87,72.9,70]
plt.pie(scores,labels=names,autopct='%1.1f%%', startangle=140, colors=['blue', 'red', 'orange'])
plt.xlabel('\n names', size=20)
plt.ylabel('scores',size=20)
plt.show()
```
